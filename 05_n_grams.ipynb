N-grams are sequences of N neighboring words or tokens in a document, used in NLP to analyze patterns and context in text.


```python
from nltk.tokenize import word_tokenize 
```

```python
x= "Natural Language Processing (NLP) is a subfield of artificial intelligence (AI) and computer science that focuses on enabling machines to understand, interpret, and generate human language. By combining computational linguistics with machine learning and deep learning, NLP bridges the gap between human communication and computer systems.NLP is widely used in applications like text translation, sentiment analysis, chatbots, and speech recognition. It powers tools such as virtual assistants (e.g., Siri, Alexa), search engines, and automated customer service systems."
```

```python
w=word_tokenize(x)
```

Types of n_grams :-
1. Unigram – Sequence of 1 word
Example: AI, is, powerful

2. Bigram – Sequence of 2 consecutive words
   Example: AI is, is powerful

3. Trigram – Sequence of 3 consecutive words
   Example: AI is powerful

```python
from nltk.collocations import BigramCollocationFinder, TrigramCollocationFinder #This line imports tools from NLTK used to find frequently occurring two-word (bigrams) and three-word (trigrams) combinations in a text.
```

# Trigram

```python
t = TrigramCollocationFinder.from_words(w)
```

```python
t.ngram_fd #This returns the frequency distribution of n-grams, showing how many times each trigram (three-word combination) appears in the text.
```

```python
t.ngram_fd.keys() #This returns all the trigram combinations (keys) found in the text without showing their frequencies.
```

# Bigram

```python
b= BigramCollocationFinder.from_words(w)
```

```python
b.ngram_fd #This returns the frequency distribution of n-grams, showing how many times each Bigram (two word combination) appears in the text.
```

```python
b.ngram_fd.keys() #This returns all the Bigram combinations (keys) found in the text without showing their frequencies.
```

# n_grams

```python
from nltk.util import ngrams
```

```python

n = ngrams(w,3) #word_tokenize

for i in n:
    print(i)
```
