Stemming is a technique used to reduce words to their base or root form by removing prefixes and suffixes (affixes), without necessarily keeping the word grammatically correct.

types of Stemming: 
1. porterstemmer
2. Regexpstemmer
3. snowballstemmer
4. Lancasterstemmer

```python
from nltk.stem import LancasterStemmer,RegexpStemmer,SnowballStemmer,PorterStemmer
```

```python
l=LancasterStemmer()
```

```python
r=RegexpStemmer('ing')
```

```python
s=SnowballStemmer('english')
```

```python
p=PorterStemmer()
```

```python
l.stem('changing')
```

```python
r.stem('changing')
```

```python
s.stem('changing')
```

```python
p.stem('changes')
```
