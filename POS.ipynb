Part of Speech(POS):POS (Part of Speech) refers to the grammatical category of a word, such as noun, verb, adjective, or adverb, based on its role in a sentence. 

```python
# Randome Text
x = 'Natural language processing has its roots in the 1950s.[2] Already in 1950, Alan Turing published an article titled "Computing Machinery and Intelligence" which proposed what is now called the Turing test as a criterion of intelligence, though at the time that was not articulated as a problem separate from artificial intelligence.'
```

```python
x 
```

```python
import nltk #import library
```

```python
nltk.download('punkt') # downloads the Punkt tokenizer model in NLTK, which is used to split text into sentences and words.
```

```python
nltk.download('punkt_tab')  #downloads the additional Punkt tokenizer data tables required by newer versions of NLTK for sentence tokenization.
```

```python
from nltk.tokenize import word_tokenize 
```

```python
w= word_tokenize(x) convert (text -> words) using  word tokenization(part of tokenization)
```

```python
nltk.download('averaged_perceptron_tagger_eng') #Downloads the English POS (Part-of-Speech) tagger model used by NLTK to label words as noun, verb, adjective, etc.
```

```python
nltk.download('averaged_perceptron_tagger') #Downloads the pre-trained POS (Part-of-Speech) tagger model in NLTK, which is used to assign grammatical tags like noun, verb, adjective, etc., to words.
```

```python
from nltk import pos_tag #(importing part of speech tags)
```

```python
p= pos_tag(w)
```
