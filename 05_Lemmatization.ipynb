Lemmatization – A process in NLP that converts a word to its base or dictionary form (lemma) while keeping the word meaningful and grammatically correct.

```python
from nltk.stem import WordNetLemmatizer #This line imports WordNetLemmatizer from NLTK, which is used to perform lemmatization (convert words to their base or dictionary form).
```

```python
wl = WordNetLemmatizer()
```

```python
wl.lemmatize('mice')
```
