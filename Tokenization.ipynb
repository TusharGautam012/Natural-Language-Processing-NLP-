Tokenization: NLP Tokenization is the process of breaking a large piece of text into smaller units like words or sentences so that a computer can understand and process it easily.

text -> words

Two types of Tokenization:
1. word tokenization
2. sentence tokenization

```python
from nltk.tokenize import sent_tokenize,word_tokenize
```

 1. sent_tokenize → splits text into sentences.
 2. word_tokenize → splits text into words.

```python
# Randome text for perform Tokenization.
text="Natural Language Processing (NLP) is a subfield of artificial intelligence (AI) and computer science that focuses on enabling machines to understand, interpret, and generate human language. By combining computational linguistics with machine learning and deep learning, NLP bridges the gap between human communication and computer systems.NLP is widely used in applications like text translation, sentiment analysis, chatbots, and speech recognition. It powers tools such as virtual assistants (e.g., Siri, Alexa), search engines, and automated customer service systems."
```

# 1. Sentence Tokenization

```python
sentence = sent_tokenize(text)
```

```python
sentence
```

```python
# Only for better showcasing.
for i in sentence:
    print(i)
    print()
```

# 2. Word Tokenization

```python
word= word_tokenize(text)
```
